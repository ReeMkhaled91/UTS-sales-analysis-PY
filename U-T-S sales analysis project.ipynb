# add new columns to answer the Q


```python
df["MONTH"] = df["Order Date"].str[0:2]
df
```

```python
df2 =df[df["MONTH"].str[0:2]=="Or"]
df2
```

```python
df = df[df["MONTH"].str[0:2]!="Or"]
```

```python
df["MONTH"] = df["MONTH"].astype(int)
```

```python
df["Quantity Ordered"] =pd.to_numeric(df["Quantity Ordered"])
df["Price Each"] =pd.to_numeric(df["Price Each"])
```

```python
df.info()
```

```python
df["TTL Sales"] = df["Quantity Ordered"]*df["Price Each"]
df.head()
```

# q1 what is the highest month for sales ? and how much was the revenue

```python
df2 =df.groupby("MONTH").sum()
df2
```

```python
df2["TTL Sales"].max()
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)
plt.bar(months,df2["TTL Sales"],color = "r")
plt.xticks(months)
plt.xlabel("months")
plt.ylabel("sales in usd$")
plt.show()
```

# Q2what is the highest city for sales ?

```python
df.head()
```

```python
def get_city (address):
    return address.split(",")[1]
def state (address):
    return address.split(",")[2].split(" ")[1]



df["city"]= df["Purchase Address"].apply(lambda x : get_city(x) +"_"+state(x))
df
```

```python
df3 =df.groupby("city").sum().reset_index()
df3
```

```python
cities = df3["city"].unique()

plt.bar(cities ,df3["TTL Sales"] ,color = "green")
plt.xticks (cities , rotation ="vertical",size =9)

plt.xlabel ("cities")
plt.ylabel ("sales in $")
plt.show()
```

# Q3 what is the best hours to display ads for customers to buy more 

```python
df["Order Date"]= pd.to_datetime(df["Order Date"])
```

```python
df["Hours"]= df["Order Date"].dt.hour
df
```

```python
hours = df.groupby("Hours").count().reset_index()
hours
```

```python
hour = hours["Hours"]
plt.plot(hour,hours["Order ID"] ,marker = "o")
plt.xticks(hour)
plt.grid()
plt.xlabel("hours")
plt.ylabel("number of order")
plt.show()
```

```python
hours.sort_values("Order ID",ascending= False).head()
```

# Q4 what is the most sold product ? and why do u think it sold most ?

```python
df_product = df.groupby("Product").sum().reset_index()
df_product
```

```python
x = df_product["Product"]
y = df_product["Quantity Ordered"]
plt.bar(x,y,color="black")
plt.xlabel("products")
plt.ylabel("quantity_ordered")
plt.xticks (x,rotation = "vertical")
plt.show()
```

```python
prices = df.groupby("Product").mean()["Price Each"]
prices
```

```python
fig , ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(x ,y ,color = "black")
ax2.plot (x,prices,"b-")
ax1.set_xlabel("products")
ax1.set_ylabel("quantity ordered")
ax2.set_ylabel("price_$")
ax1.set_xticklabels(x,rotation="vertical")
plt.show()
```
